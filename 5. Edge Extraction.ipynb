# Edge Extraction

#### Sobel Filter

```python
#Importing the necessary package
from skimage.io import imread, imshow
from skimage.filters import sobel_h, sobel_v
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Read the image
taj = imread("./image/taj.jpg", as_gray = True)
imshow(taj)
```

```python
# Edge extraction
taj_gray_sobelh = sobel_h(taj)   # Horizontal extraction
taj_gray_sobelv = sobel_v(taj)   # Vertical extraction
```

```python
# Plotting images
plt.figure(1, figsize = (15, 25))

plt.subplot(131), imshow(taj)
plt.title("Original Image")

plt.subplot(132), imshow(taj_gray_sobelh)
plt.title("Horizontal Edge Image")

plt.subplot(133), imshow(taj_gray_sobelv)
plt.title("Vertical Edge Image")

plt.show()
```

#### Prewitt Filter

```python
# Import the necessary package
from skimage.filters import prewitt_h, prewitt_v
```

```python
# Edge Extraction
taj_gray_prewitth = prewitt_h(taj)    # Horizontal Edge
taj_gray_prewittv = prewitt_v(taj)    # Veritical Edge
```

```python
# Plot the images
plt.figure(1, figsize = (15, 25))

plt.subplot(131), imshow(taj)
plt.title("Original Image")

plt.subplot(132), imshow(taj_gray_prewitth)
plt.title("Horizontal Image")

plt.subplot(133), imshow(taj_gray_prewittv)
plt.title("Vertical Image")

plt.show()
```
