# Image Rotation & Flipping

```python
# Importing the necessary packages
from skimage.transform import rotate
from skimage.io import imread, imshow
from numpy import fliplr, flipud
import matplotlib.pyplot as plt
```

```python
# Read the image
taj = imread("./image/taj.jpg", as_gray = False)
imshow(taj)
```

```python
# Rotate the image by 40 degree
taj_rotated = rotate(taj, angle = 40)
imshow(taj_rotated)
```

```python
# Flipping the image from left to right
taj_fliped = fliplr(taj)
imshow(taj_fliped)
```

```python
# Flipped the image upside down
taj_fliped_ud = flipud(taj)
imshow(taj_fliped_ud)
```
