# Image Intensity

```python
# Import the necessary packages
from skimage import exposure
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

```python
# Read the image
taj = imread("./image/taj.jpg", as_gray = False)
imshow(taj)
```

```python
# Adjusting the brightness of the image
taj_bright = exposure.adjust_gamma(taj, gamma = 0.25, gain = 1)
taj_dark = exposure.adjust_gamma(taj, gamma = 2.5, gain = 1)
taj_darkest = exposure.adjust_gamma(taj, gamma = 4, gain = 1)
```

```python
# Plot the images with adjusted brightness
plt.figure(1, figsize = (10,10))

plt.subplot(221), imshow(taj_dark)
plt.title("Darker Image")

plt.subplot(222), imshow(taj_darkest)
plt.title("Darkest Image")

plt.show()
```

# Image Cropping

```python
# Cropping of image in required dimensions
taj_img_crop_1 = taj[50:(taj.shape[0]-10), 50:(taj.shape[1]-10)]
taj_img_crop_2 = taj[60:(taj.shape[0]-20), 60:(taj.shape[1]-30)]
```

```python
# Plot the cropped images
plt.figure(1, figsize = (10,15))

plt.subplot(131), imshow(taj)
plt.title("Original Image")

plt.subplot(132), imshow(taj_img_crop_1)
plt.title("Cropped Image")

plt.subplot(133), imshow(taj_img_crop_2)
plt.title("2nd Cropped Image")

plt.show()
```
