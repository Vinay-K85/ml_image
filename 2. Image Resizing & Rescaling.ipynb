# Image Resizing & Rescaling

```python
# Importing the necessary package
from skimage.transform import resize, rescale
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

```python
# Read the image
taj = imread("./image/taj.jpg", as_gray = True)
imshow(taj)
```

```python
# Original image
print("Original dimension: ", taj.shape)
```

```python
# Resize the image
taj_resize = resize(taj, (120,120))
print("Resized dimension: ", taj_resize.shape)
```

```python
# Rescaling image @ 50%
taj_rescaled = rescale(taj, scale = 0.5)
print("Rescaled dimension: ", taj_rescaled.shape)
```

```python
# New rescaling of image
taj_rescaled_1 = rescale(taj, scale = 0.75)
print("New rescaled dimension to 75%: ", taj_rescaled_1.shape)
```

```python
# Plot images after rescaling and resizing
plt.figure(1, figsize = (10,10))
plt.subplot(221), imshow(taj)
plt.title("Original Image")

plt.subplot(222), imshow(taj_resize)
plt.title("Resized Image")

plt.subplot(223), imshow(taj_rescaled)
plt.title("Rescaled Image")

plt.subplot(224), imshow(taj_rescaled_1)
plt.title("New Rescaled Image")

plt.show()
```
