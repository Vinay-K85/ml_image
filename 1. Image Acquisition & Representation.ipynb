# Image Acquisition & Representation

```python
# Import the necessary package for reading image
from skimage.io import imread, imshow
```

```python
# Reading the image
taj_img_color = imread("./image/taj.jpg", as_gray = False)
print("The shape of the colored image is: ", taj_img_color.shape)
print(taj_img_color)
imshow(taj_img_color)
```

```python
# Reading the image in gray color
taj_img_gray = imread("./image/taj.jpg", as_gray = True)
print("The shape of of the gray image is: ", taj_img_gray.shape)
print(taj_img_gray)
imshow(taj_img_gray)
```

```python
# Using function to display image output in different colors
# Import the necessary packages
import matplotlib.pyplot as plt
from skimage.color import rgb2gray, rgb2hsv
```

```python
taj = imread("./image/taj.jpg")
taj_gray = rgb2gray(taj)  # function will display image gray
taj_hsv = rgb2hsv(taj)  # function will disply image in hsv

plt.figure(1, figsize = (30,10))
plt.subplot(131), imshow(taj)
plt.title("Original RGB Format")

plt.subplot(132), imshow(taj_gray)
plt.title("Gray Scale Format")

plt.subplot(133), imshow(taj_hsv)
plt.title("HSV Format")

plt.show()
```
